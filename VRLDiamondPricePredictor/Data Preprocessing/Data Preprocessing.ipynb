```python
def Combine_df(path1,path2):
    '''
    It takes 2 paths and concat the corresp df and return
    '''
    return pd.concat([pd.read_csv(path1), pd.read_csv(path2)])

```

# Working on Messurements


```python
# Messesurements has 3 features as 1 str i.e. Depth, Height, Width
# Two types of division 
    #1. Normal:  Depth-Height×Width  
    #2. Errors in B/w:
    #      a) Err1: DepthXHeightXWidth    [almost 90% have this error]
    #      b) Err2: Depth-HeightXWidth    [only few i.e.round category]
```

```python
df_comp.Messurements
```

```python
df_comp[df_comp.Shape.values=="ROUND"].Messurements
```

```python
# All the Error Terms

df_comp.Messurements.apply(lambda x : x if "X" in x else np.nan).dropna()
```

```python
def Error1(x):
    s = map(float,x.split("X"))
    return tuple(s)

def Error2(x):
    s = "X".join(x.split("-"))
    return Error1(s)

def Normal(x):
    s = x.split("-")
    s1 = s[1].split("×")
    s= s[:1]
    s.extend(s1)
    s = map(float,s)
    return tuple(s)
    
```

```python
for i in Error1("5.10X5.06X3.15"):
    print(i)
```

```python
for i in Error2("3.85-3.84X2.42"):
    print(i)
```

```python
for i in Normal("4.36-4.40×2.59"):
    print(i)
```

```python
d_mess={"Depth":[] ,  "Height":[], "Width":[]}
for x in df_comp.Messurements.values:
    if "X" in x and "-" in x:
        # If both X and - 
        dat=Error2(x)
    
    elif "X" in x:
        # Only X
        dat=Error1(x)
        
    else:
        # Normal Case :  Depth-Height×Width
        if "mm" in x:
            x=(x[:-3].replace(" ","")).replace("–","-").replace("x","×")  # Normal exception
            x= x.strip()
        dat=Normal(x)
    
    #
    print(dat)
    d_mess["Depth"].append(dat[0])
    d_mess["Height"].append(dat[1])
    d_mess["Width"].append(dat[2])
    
```
