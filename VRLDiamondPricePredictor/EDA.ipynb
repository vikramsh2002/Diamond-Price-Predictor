# Reverse Maps to Find Categorical Statistical Summary

```python
def ReverseDictionary(dct):
    rev = pd.Series(dct.keys(),index = dct.values()).T
    return rev.to_dict()
```

```python
ReverseDictionary(maps["Clarity"])
```

```python
df
```

```python
list(maps.keys())[:-1]
```

```python
cat_feat= df[["Clarity","Cut","Colour","Polish","Symmetry","Shape","Fluorescence"]]
```

```python
# Reverse  mapp
cat_feat.Clarity=cat_feat.Clarity.map(ReverseDictionary(maps["Clarity"]))
cat_feat.Cut=cat_feat.Cut.map(ReverseDictionary(maps["Goodness"]))
cat_feat.Polish=cat_feat.Polish.map(ReverseDictionary(maps["Goodness"]))
cat_feat.Symmetry=cat_feat.Symmetry.map(ReverseDictionary(maps["Goodness"]))
cat_feat.Fluorescence=cat_feat.Fluorescence.map(ReverseDictionary(maps["Fluorescence"]))
cat_feat.Shape=cat_feat.Shape.map(ReverseDictionary(maps["Shapes"]))
cat_feat.Colour=cat_feat.Colour.map(ReverseDictionary(maps["Color"]))


```

```python
cat_feat
```

```python
cat_feat["Price"] = df.Price
```
